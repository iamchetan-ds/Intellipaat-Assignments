```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
# Load cars data as dataframe using pandas and create a bar plot between number of cylinders and frequency of cars with that many number of cylinders
  #  - Set xlabel as Number of cylinders
  #  - Set ylabel as Frequency of cars
  #  - Draw a bar plot
    
freq_table = pd.crosstab(index= df['cyl'],
           columns= 'count').reset_index()

sns.countplot(x = df['cyl'])
plt.yticks(ticks = freq_table['count'])
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.show()
```

```python
# Write code to load data from cars and print a bar graph of count of columns with null values

def null_value_plot(df):
    null_df = df.isnull().sum().to_frame().reset_index()
    null_df.columns = ['column_name', 'null_count']
    
    null_df = null_df.sort_values(by = 'null_count', ascending = False)
    
    plt.rcParams["figure.figsize"] = [10, 6]
    plt.rcParams["figure.autolayout"] = True

    fig, ax = plt.subplots()             
    null_df['null_count'].plot(ax=ax, kind='bar', xlabel='numbers', ylabel='frequency')
    plt.show()
    
null_value_plot(df)
```

```python
# Use the 'mpg' (Miles Per Gallon column) and draw a histogram:
  #  - Set xlabel: Miles per gallon
  #  - Set ylabel: Frequency
  #  - Set title as Miles Per Gallon Histogram
  #  - Use mpg column to generate a histogram

sns.histplot(x =df['mpg'])
plt.xlabel('Miles per gallon')
plt.ylabel('Frequency')
plt.title('Miles Per Gallon Histogram')
plt.show()
```

```python
# Draw a boxplot on the card dataframes hp column:
# Set xlabel: Car Horsepower
# Set title as Boxplot for car horsepower
# Use hp column to generate a boxplot

sns.boxenplot(x =df['hp'])
plt.xlabel('Car Horsepower')
plt.title('Boxplot for car horsepower')
plt.show()
```
