```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Start off by importing the cars.csv file in the jupyter notebook.
df = pd.read_csv('cars.csv')
```

```python
df.head()
```

```python
# Generate a line plot graph for the column ‘model’ and ‘hp’:

sns.lineplot(data = df, x = 'model', y = 'hp')

plt.xlabel(' Models of the cars')
plt.xticks(rotation=90)
plt.ylabel('Horse-Power of Cars')
plt.title('Model Names vs HorsePow')
```

```python
# Generate a bar plot graph for the columns ‘carbs’ and ‘gear’

sns.barplot(data = df, x = 'carb', y= 'gear')

plt.xlabel(' Number of carburetors')
plt.ylabel('Number of forward gears')
plt.title('carbs vs gear')
plt.show()
```

```python
# Plot a histogram for the column ‘wt’:
sns.histplot(df['wt'], bins=30)

plt.xlabel('Weight of the cars')
plt.ylabel('Count')
plt.title('Histogram for the weight')
plt.show()
```

```python
# Plot a pie chart for columns: ‘cyl’ and ‘model’ form the mtcars.csv data frame.
ct = pd.crosstab(index = df['cyl'],
           columns='Count').reset_index()

plt.pie(x = ct['Count'],
       labels=ct['cyl'])
plt.show()
```

```python
# Plot the area chart for the columns ‘am’ and ‘carb’:
 #   - Set the ‘am’ on the x-axis
 #   - Set the ‘carb’ on the y-axis
 #   - Provide the x-axis label as Transmission
 #   - Provide the y-axis label as Number of carburetors
 #   - Provide the title as Transmission vs Number of carburetors

plt.stackplot(df['am'], df['carb'])
plt.xlabel('Transmission')
plt.ylabel('Number of carburetors')
plt.title('Transmission vs Number of carburetors')
plt.show()
```
