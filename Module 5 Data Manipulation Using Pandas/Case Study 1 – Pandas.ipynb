```python
import pandas as pd
```

```python
# Write a function that takes start and end of a range returns a pandas series object containing numbers within that range.
def range_series(start = 1, stop = 10):
    arr = list(range(start, stop+1))
    return pd.Series(arr)

print('range_series() : ', range_series())
print('range_series(5) : ', range_series(5))
print('range_series(5,15) : ', range_series(5,15))
```

```python
# Create a function that takes in two lists named keys and values as arguments
# Keys would be strings and contain n string values,
# Values would be a list containing n lists.
# The methods should return a new pandas DataFrame with keys as column names and values as their corresponding values

def create_dataframe(keys, values):
    return pd.DataFrame(data = values,
                       columns = keys)
create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]])
```

```python
# Create a function that concatenates two DataFrames. 
# Use a previously created function to create two DataFrames and 
# pass them as parameters Make sure that the indexes are reset before returning.

def concat_dataframes(df1, df2):
    frames = [df1, df2]
    return pd.concat(frames)


a = create_dataframe(["One", "Two"], [["X", "Y"], ["A", "B"]])
b = create_dataframe(["Three", "Four"], [["X", "Y"], ["C", "D"]])

concat_dataframes(a, b)
```

```python
# Write code to load data from cars.csv into a dataframe and print its details. 
# Details like: 'count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max'

df = pd.read_csv('cars.csv')
df.describe()
```

```python
# Write a method that will take a column name as argument and 
# return the name of the column with which the given column has the highest correlation.

def get_max_correlated_column(df, column):
    """
    takes dataframe and a column name as a input
    return the highest correlated column name with input column
    """
    col = df.corr()[column].to_frame().reset_index().sort_values(by = 'mpg', ascending=False).iloc[1,0]
    val = df.corr()[column].to_frame().reset_index().sort_values(by = 'mpg', ascending=False).iloc[1,1]
    return print('( column =',col, ',' 'pearsonr =',val,')')

get_max_correlated_column(df, 'mpg')
```
